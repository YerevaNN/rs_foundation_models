```python
# UCMerced

import matplotlib.pyplot as plt
from scipy.integrate import trapz

# Data points
x_values = [0, 1/8, 1/4, 1/2, 1]

y_values_ibot_milaid = [0, 0.117, 0.510, 0.983, 0.988]
y_values_gfm = [0, 0.433, 0.781, 0.914, 0.948]
y_values_ibot_imagenet = [0, 0.652, 0.905, 0.983, 0.981]
# y_values_satlas = [0, 0.276, 0.367, 0.398, 0.414]


# Sort together by x-values
# sorted_data = sorted(zip(x_values, y_values))
# x_values, y_values = zip(*sorted_data)

# Calculate AUC using trapezoidal rule
auc_ibot_imagenet = trapz(y_values_ibot_imagenet, x_values)
auc_ibot_milaid = trapz(y_values_ibot_milaid, x_values)
auc_gfm = trapz(y_values_gfm, x_values)

# auc_satlas = trapz(y_values_satlas, x_values)

plt.figure(figsize=(4, 4))

# Plot the curve and points
plt.plot(x_values, y_values_ibot_imagenet, label=f"iBOT-IMAGENET ({auc_ibot_imagenet:.3f})", marker='o')
plt.plot(x_values, y_values_gfm, label=f"GFM ({auc_gfm:.3f})", marker='o')
plt.plot(x_values, y_values_ibot_milaid, label=f"iBOT-MILLIONAID ({auc_ibot_milaid:.3f})", marker='o')

# plt.plot(x_values, y_values_satlas, label=f"SATLAS ({auc_satlas:.3f})", marker='o')
# plt.scatter(x_values, y_values, label="Data Points")

# Add labels and title
plt.xlabel("Scale")
plt.ylabel("Micro-F1")
plt.title("UC Merced")

# Show AUC value
# plt.text(0.7, 0.1, f"AUC: {auc:.3f}", ha="center", va="center", color='C0')
# plt.text(0.2, 0.4, f"AUC: {auc_aug:.3f}", ha="center", va="center", color='C1')

plt.legend(bbox_to_anchor=(0.05, -0.2), loc='upper left') 

# plt.legend(loc='outer')
plt.grid(True)
plt.ylim(0, 1)
plt.xlim(0, 1)
plt.savefig('classification_ucmerced.pdf', bbox_inches='tight') 
plt.show()


```
